```python
#PhÆ°Æ¡ng trÃ¬nh báº­c 2 má»™t áº©n cÃ³ dáº¡ng ğ‘ğ‘¥ + ğ‘ğ‘¥ + ğ‘ = 0 vá»›i ğ‘ â‰  0. Viáº¿t chÆ°Æ¡ng trÃ¬nh yÃªu cáº§u ngÆ°á»i dÃ¹ng nháº­p há»‡ sá»‘ ğ‘,ğ‘,ğ‘. TÃ­nh toÃ¡n vÃ  xuáº¥t ra mÃ n hÃ¬nh nghiá»‡m cá»§a bÃ i toÃ¡n
import math
a=float(input("Vui lÃ²ng nháº­p sá»‘ A: "))
b=float(input("Vui lÃ²ng nháº­p sá»‘ B: "))
c=float(input("Vui lÃ²ng nháº­p sá»‘ C: "))
delta=b**2 - (4*a*c)
if delta < 0:
    print("PhÆ°Æ¡ng trÃ¬nh vÃ´ nghiá»‡m!")
elif delta > 0:
    print("PhÆ°Æ¡ng trÃ¬nh cÃ³ 2 nghiá»‡m phÃ¢n biá»‡t: ")
    x1 = (-b + math.sqrt(delta))/(2*a)
    x2 = (-b - math.sqrt(delta))/(2*a)
    print("x1 =",x1 , "x2 =",x2)
elif delta == 0:
    z1 = -b/2*a
    print(" PhÆ°Æ¡ng trÃ¬nh cÃ³ nghiá»‡m kÃ©p x1 = x2 = ", z1)
```

```python
#Viáº¿t chÆ°Æ¡ng trÃ¬nh cho ngÆ°á»i dÃ¹ng nháº­p vÃ o 1 sá»‘ nguyÃªn dÆ°Æ¡ng n. Xuáº¥t cÃ¡c thá»«a sá»‘ nguyÃªn tá»‘ cá»§a n
def phanTichSoNguyen(n):
    i = 2;
    listNumbers = [];
    # phÃ¢n tÃ­ch
    while (n > 1):
        if (n % i == 0):
            n = int(n / i);
            listNumbers.append(i);
        else:
            i = i + 1;
    if (len(listNumbers) == 0):
        listNumbers.append(n);
    return listNumbers;
 
n = int(input("Nháº­p sá»‘ nguyÃªn dÆ°Æ¡ng n = "));
listNumbers = phanTichSoNguyen(n);
size = len(listNumbers);
sb = "";
for i in range(0, size - 1):
    sb = sb + str(listNumbers[i]) + " x ";
sb = sb + str(listNumbers[size-1]);
print("Káº¿t quáº£:", n, "=", sb);
```

```python
#Viáº¿t chÆ°Æ¡ng trÃ¬nh sáº¯p xáº¿p cÃ¡c pháº§n tá»­ trong danh sÃ¡ch cÃ¡c sá»‘ nguyÃªn tÄƒng dáº§n báº±ng 
Interchange sort, Bubble sort vÃ  Binary sort
Interchange sort, Bubble sort vÃ  Binary sort
a=int(input("Vui lÃ²ng chá»n thuáº­t toÃ¡n: 1: Interchange sort || 2: Bubble sort || 3: Binary sort"))
if a == 1:
    arr = [64, 34, 25, 12, 22, 11, 90]
    n = len(arr)
    for i in range(n - 1):
        for j in range(i + 1, n):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
    print("Danh sÃ¡ch sau khi sáº¯p xáº¿p báº±ng Interchange sort:")
    print(arr)
elif a == 2:
    arr = [64, 34, 25, 12, 22, 11, 90]
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    print("Danh sÃ¡ch sau khi sáº¯p xáº¿p báº±ng Bubble sort:")
    print(arr)
elif a == 3:
    arr = [64, 34, 25, 12, 22, 11, 90]
    for i in range(1, len(arr)):
        key = arr[i]
        left, right = 0, i - 1
        while left <= right:
            mid = (left + right) // 2
            if arr[mid] > key:
                right = mid - 1
            else:
                left = mid + 1
        arr[:] = arr[:left] + [key] + arr[left:i] + arr[i + 1:]
    print("Danh sÃ¡ch sau khi sáº¯p xáº¿p báº±ng Binary sort:")
    print(arr)
else:
    print("Báº¡n nháº­p sai má»i nháº­p láº¡i!")
```

```python
#Viáº¿t chÆ°Æ¡ng trÃ¬nh tÃ¬m kiáº¿m 1 sá»‘ trong danh sÃ¡ch cÃ¡c sá»‘ nguyÃªn báº±ng tÃ¬m kiáº¿m tuáº§n tá»± vÃ  tÃ¬m kiáº¿m nhá»‹ phÃ¢n
a=int(input("Vui lÃ²ng chá»n thuáº­t toÃ¡n: 1: TÃ¬m kiáº¿m tuáº§n tá»± || 2: TÃ¬m kiáº¿m nhá»‹ phÃ¢n"))
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid + 2
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
if a == 1:
    arr = list(map(int, input("Nháº­p dÃ£y sá»‘ nguyÃªn, cÃ¡ch nhau báº±ng dáº¥u cÃ¡ch: ").split()))
    target = int(input("Nháº­p sá»‘ cáº§n tÃ¬m kiáº¿m: "))
    result = linear_search(arr, target)
    if result != -1:
        print("Sá»‘", target, "Ä‘Æ°á»£c tÃ¬m tháº¥y táº¡i chá»‰ sá»‘", result)
    else:
        print("Sá»‘", target, "khÃ´ng cÃ³ trong danh sÃ¡ch.")
elif a == 2:
    arr = sorted(list(map(int, input("Nháº­p dÃ£y sá»‘ nguyÃªn, cÃ¡ch nhau báº±ng dáº¥u cÃ¡ch: ").split())))
    target = int(input("Nháº­p sá»‘ cáº§n tÃ¬m kiáº¿m: "))
    result = binary_search(arr, target)
    if result != -1:
        print("Sá»‘", target, "Ä‘Æ°á»£c tÃ¬m tháº¥y táº¡i chá»‰ sá»‘", result)
    else:
        print("Sá»‘", target, "khÃ´ng cÃ³ trong danh sÃ¡ch.")
```

```python
#Trong pháº§n má»m Excel, cÃ¡c cá»™t sáº½ Ä‘Æ°á»£c Ä‘Ã¡nh chá»‰ sá»‘ (index) báº±ng cÃ¡c kÃ½ tá»± latin in hoa (A Ã  Z). VÃ­ dá»¥ â€¢ A: 1 â€¢ B: 2 â€¢ C: 3 â€¢ â€¦ â€¢ Z: 26 â€¢ AA: 27 â€¢ AB: 28 Viáº¿t 1 chÆ°Æ¡ng trÃ¬nh nháº­p sá»‘ nguyÃªn ğ‘› vÃ  Ä‘á»•i nÃ³ qua tÃªn cá»§a má»™t cá»™t trong excel
n = int(input("Nháº­p sá»‘ nguyÃªn n: "))
def index_to_excel_column(n):
    result = ""
    while n > 0:
        remainder = (n - 1) % 26
        result = chr(65 + remainder) + result
        n = (n - 1) // 26
    return result
column_name = index_to_excel_column(n)
print(f"TÃªn cá»™t trong Excel tÆ°Æ¡ng á»©ng vá»›i sá»‘ nguyÃªn {n} lÃ : {column_name}")
```
